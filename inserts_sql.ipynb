```python
grau_academico = {
    1: "Bacharelado",
    2: "Licenciatura",
    3: "Tecnólogo",
}
```

```python
df3 = pd.read_csv('/content/MICRODADOS_CADASTRO_CURSOS_2023.CSV', sep=';')

df4 = df3[['NO_CURSO','NO_CINE_AREA_GERAL', 'TP_GRAU_ACADEMICO']]

df5 = df4.dropna(how="any").drop_duplicates()

df5.columns = ['CURSO', 'AREA', 'GRAU']

df5['GRAU'] = df5['GRAU'].map(grau_academico)

df5
```

```python
ids = random.sample(range(0, len(df5)), 80)

tuplas = [tuple(df5.iloc[i]) for i in ids]

query = "INSERT INTO cursos (nome, area, modalidade, grau_academico, mensalidade, vagas_ofertadas, vagas_preenchidas, vagas_restantes, candidatos, ies_id) VALUES \n"

for _ in range(350):
  curso = random.choice(tuplas)
  nome = curso[0]
  area = curso[1]
  modalidade = random.choice(["Presencial", "EaD"])
  grau_academico = curso[2]
  mensalidade = random.randint(1000, 10000)
  vagas_ofertadas = random.randint(10, 70)
  vagas_preenchidas = random.randint(0, vagas_ofertadas)
  vagas_restantes = vagas_ofertadas - vagas_preenchidas
  candidatos = random.randint(100, 1000)
  ies_id = random.randint(1,103)

  query += f"('{nome}', '{area}', '{modalidade}', '{grau_academico}', {mensalidade}, {vagas_ofertadas}, {vagas_preenchidas}, {vagas_restantes}, {candidatos}, {ies_id}),\n"

query = query[:-2] + ";"

print(query)

with open('insert_cursos.sql', 'w') as f:
  f.write(query)
```

```python
status_matricula = [
    "Ativo",
    "Formado",
    "Evasão"
]

tipo_bolsas = [
    "Prouni",
    "FIES",
    "Bolsa Institucional"
]
```

```python
query = "INSERT INTO matriculas (ano_ingresso, ano_conclusao, status, bolsista, tipo_bolsa, percentual_bolsa, cursos_id, discentes_id) VALUES \n"

for _ in range(2000):
  ano_ingresso = random.randint(2016, 2024)
  status = random.choice(status_matricula)
  if status == "Evasão":
    ano_conclusao = random.randint(ano_ingresso, min(ano_ingresso+4, 2024))
  else:
    ano_conclusao = ano_ingresso+4
    if status == "Formado" and ano_conclusao >= 2024:
      status = "Ativo"
    elif status == "Ativo" and ano_conclusao < 2024:
      status = "Formado"
  bolsista = random.choice([True, False])
  if bolsista:
    tipo_bolsa = random.choice(tipo_bolsas)
    if tipo_bolsa in ["Prouni", "FIES"]:
      percentual_bolsa = random.randint(5, 10)*10
    else:
      percentual_bolsa = random.randint(1, 10)*10

  cursos_id = random.randint(1,353)
  discentes_id = random.randint(1,1003)

  if bolsista:
    query += f"({ano_ingresso}, {ano_conclusao}, '{status}', TRUE, '{tipo_bolsa}', {percentual_bolsa}, {cursos_id}, {discentes_id}),\n"
  else:
    query += f"({ano_ingresso}, {ano_conclusao}, '{status}', FALSE, NULL, NULL, {cursos_id}, {discentes_id}),\n"

query = query[:-2] + ";"

print(query)

with open('insert_matriculas.sql', 'w') as f:
  f.write(query)
```
